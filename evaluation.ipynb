```python
import pandas as pd

# Load CSV file into a pandas DataFrame
df = pd.read_csv('my_final_dataset.csv')
```

```python
df.head()
```

```python
from transformers import GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("EleutherAI/gpt-neo-2.7B")
```

```python
df["java_candidate_tok"]=df["java_candidate"].apply(tokenizer.encode)

df["c#_candidate_tok"]=df["c#_candidate"].apply(tokenizer.encode)
```

```python
df
```

```python
def bleu_score(lst1,lst2):
    lst1=set(eval(lst1))
    lst2=set(lst2)
    shared_items = lst1 & lst2 
    return (len(shared_items)/len(lst2))
```

```python
res = []
for i in range(32):
    res.append(bleu_score(df.at[i, "c#_tok"], df.at[i, "c#_candidate_tok"]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# Calculate mean and standard deviation
mu = np.mean(res)
sigma = np.std(res)

# Plot histogram with normal distribution curve
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(res, bins=8, density=True, alpha=0.5, color='green')
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
ax.plot(x, norm.pdf(x, mu, sigma), color='red', linewidth=2)
ax.set_xlabel('BLEU Scores for the 32 testing examples')
ax.set_ylabel('Desnity of occurence of BLEU scores')
ax.set_title('Histogram of 32 BLEU scores for JAVA to C# with Normal Distribution Curve')
plt.show()

print(" *************************************************************************\nThe mean vlaue of BLEU score for Java to c# translation is: \n \n",sum(res)/len(res))
```

```python
def rouge_score(lst1,lst2):
    lst1=set(eval(lst1))
    lst2=set(lst2)
    shared_items = lst1 & lst2 
    return (len(shared_items)/len(lst1))
```

```python
res3 = []
for i in range(32):
    res3.append(rouge_score(df.at[i, "c#_tok"], df.at[i, "c#_candidate_tok"]))
```
